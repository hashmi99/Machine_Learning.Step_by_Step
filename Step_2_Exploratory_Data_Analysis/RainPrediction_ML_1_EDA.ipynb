**Importing Libraries**

```python
import numpy as  np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Read Dataset through Pandas**

```python
df = pd.read_csv("/content/weatherAUS.csv")
df
```

**Separating/ Segregating Features**

- Numerical Features

- Categorical Features

```python
Numerical_Features = [feature for feature in df.columns if df[feature].dtype != 'object' ]
Categorical_Features = [feature for feature in df.columns if df[feature].dtype == 'object' ]
```

**From Extracted Numerical Features**

- Descrete Features

- Continuous Features

```python
Descrete_Feature = [feature for feature in Numerical_Features if len(df [feature].unique ())<= 25 ]
Continuous_Feature = [feature for feature in Numerical_Features if feature not in Descrete_Feature ]
```

```python
print ("\nNumerical Features are......  \n\n", Numerical_Features)
print ("\nCategorical_Features are......  \n\n", Categorical_Features)
print ("\nDescrete_Features are......  \n\n", Descrete_Feature)
print ("\nContinuous_Feature are......  \n\n", Continuous_Feature)
```

**Assessing the Distribution**

```python
for feature in Numerical_Features:
  sns.distplot(df[feature])
  plt.show()
```

**Removing KDE Curve**

```python
for feature in Numerical_Features:
  sns.distplot(df[feature], kde = False)
  plt.show()
```

```python
for feature in Numerical_Features:
  sns.distplot(df[feature], bins = 31, kde = False)
  plt.show()
```

**Analysing Missing Values**



```python
missing_values = df.isnull (). sum ()
print (missing_values)
```

```python
missing_values.sort_values (ascending = False, inplace = True)
missing_values = missing_values.to_frame ()
missing_values.columns = [" Missing Values"]
missing_values.index.names = ['Feature Name']
missing_values ["Feature Name"] = missing_values.index
sns.barplot (x= "Feature Name" , y = " Missing Values" , data = missing_values)
plt.xticks (rotation = 90)
plt.show ()
```
